```python
# importing necessary libraries for data preprocessing:

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
data = pd.read_csv("C:/Users/tkjay/Desktop/goog.csv")
```

```python
data.tail()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data_train = data[data['Date']<'2019-01-01'].copy()
data_train
```

```python
data_test = data[data['Date']>'2019-01-01'].copy()
data_test
```

```python
# Creating the actual training data by dropping the unnecessary columns

training_data = data_train.drop(['Date','Adj Close'],axis = 1)
training_data.head()
```

```python
# importing and executing StandardScaler to make the data between 0 and 1:

scaler = MinMaxScaler()
training_data = scaler.fit_transform(training_data)
training_data
```

```python
x_train = []
y_train = []

# Now we need to put in the first 60 days worth of data in x_train

for i in range(60 , training_data.shape[0]):
    x_train.append(training_data[i-60:i])
    y_train.append(training_data[i,0])
```

```python
x_train, y_train = np.array(x_train),np.array(y_train)
```

```python
x_train.shape
```

```python
y_train.shape
```

### Builidng LTSM

```python
#importing keras

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,LSTM,Dropout
```

```python
regressior = Sequential()
regressior.add(LSTM(units = 100,activation = 'relu',return_sequences = True,input_shape = (60,5)))
```

```python
regressior.add(Dropout(0.2))
```

```python
regressior = Sequential()
regressior.add(LSTM(units = 100,activation = 'relu',return_sequences = True,input_shape = (60,5)))
regressior.add(Dropout(0.2))


regressior.add(LSTM(units = 120,activation = 'relu',return_sequences = True,))
regressior.add(Dropout(0.3))

regressior.add(LSTM(units = 150,activation = 'relu',return_sequences = True,))
regressior.add(Dropout(0.4))


regressior.add(LSTM(units = 180,activation = 'relu'))
regressior.add(Dropout(0.5))


regressior.add(Dense(units = 1))
```

```python
regressior.summary()
```

```python
regressior.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
regressior.fit(x_train,y_train,epochs = 10,batch_size = 32)
```
